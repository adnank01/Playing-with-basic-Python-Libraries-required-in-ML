```python
import numpy as np
from matplotlib import pyplot as plt
```

Line Plot

```python
x=np.arange(1,11)
x
```

```python
y= 2*x
y
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.plot(x,y, color="g", linestyle=":", linewidth=4)
plt.show()
```

```python
y2=4*x
y2
```

```python
plt.plot(x,y,linestyle="-.")
plt.plot(x,y2, color="g", linestyle=":", linewidth=2)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1) 
plt.plot(x,y,linestyle="-.")
                                                ##parameters of subplot : (row,column,index of that graph)

plt.subplot(2,1,2)
plt.plot(x,y2, color="g", linestyle=":", linewidth=2)


plt.show()
```

```python
x=np.arange(1,11)
y= 2*x
y2=4*x
plt.subplot(1,2,1) 
plt.plot(x,y,linestyle="-.")
                                                ##parameters of subplot : (row,column,index of that graph)

plt.subplot(1,2,2)
plt.plot(x,y2, color="g", linestyle=":", linewidth=2)

plt.show()
```

Bar plots and horizontal bar plots 

```python
student={'bob':97,'Rock':80,'Mike': 70}
x=list(student.keys())
y=list(student.values())
plt.bar(x,y)
plt.title("Bar Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
plt.barh(x,y,color='green')
plt.title("Bar Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

Scatter Plot

```python
xs=np.arange(10,100,10)
ys=[4,6,8,9,5,7,3,1,2]
plt.scatter(xs,ys)
plt.show()
```

```python
plt.scatter(xs,ys,marker='*', c='cyan',s=200)
plt.show()
```

```python
a=[4,7,9,8,3,6,1,2,5]
plt.scatter(xs,ys,marker='*', c='green',s=50)
plt.scatter(xs,a, c='black',s=50)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(xs,ys,marker='*', c='green',s=50)
plt.subplot(1,2,2)
plt.scatter(xs,a, c='black',s=50)
plt.show()
```

Histogram and Box Plot

```python
data=[1,1,3,3,3,3,4,6,6,6,6,6,6,7,7,9,9,9] #here,no. of times that particular value has occurred
plt.hist(data)
plt.show()
```

```python
import pandas as pd
```

```python
iris=pd.read_csv('iris.csv')
iris.head()
```

```python
plt.hist(iris['Petal.Length'],color='r', bins=20)
plt.show()
```

Box Plot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,6,5,4,3]

data=list([one,two,three])
plt.boxplot(data)         ##gives us median value, max and min value of an array of values
plt.show()
```

Violin plot

```python
plt.violinplot(data,showmedians='True')         ##gives us median value, max and min value of an array of values
plt.show()
```

Pie-Chart

```python
iris_species=iris[['Species']].value_counts().to_dict()
iris_species
```

```python
plt.pie(list(iris_species.values()),labels=list(iris_species.keys()))
plt.show()
```

```python
plt.pie(list(iris_species.values()),labels=list(iris_species.keys()), autopct='%0.1f%%',colors=['Cyan','Grey','Violet'])
plt.show()
```

Dough Nut Chart

```python
plt.pie(list(iris_species.values()),labels=list(iris_species.keys()), autopct='%0.1f%%',colors=['Cyan','Grey','Violet'], radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
